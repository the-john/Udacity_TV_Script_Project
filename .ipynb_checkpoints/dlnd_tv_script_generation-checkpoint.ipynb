### Lookup Table
To create a word embedding, you first need to transform the words to ids.  In this function, create two dictionaries:
- Dictionary to go from the words to an id, we'll call `vocab_to_int`
- Dictionary to go from the id to word, we'll call `int_to_vocab`

Return these dictionaries in the following **tuple** `(vocab_to_int, int_to_vocab)`

````python
import problem_unittests as tests
from collections import Counter

def create_lookup_tables(text):
    """
    Create lookup tables for vocabulary
    :param text: The text of tv scripts split into words
    :return: A tuple of dicts (vocab_to_int, int_to_vocab)
    """
    # TODO: Implement Function
    
    word_counts = Counter(text)
    # Sort words starting with the most frequent words occuring in the text
    sorted_vocab = sorted(word_counts, key=word_counts.get, reverse=True)
    
    # Now, create the two needed dictionaries
    int_to_vocab = {ii: word for ii, word in enumerate(sorted_vocab)}
    vocab_to_int = {word: ii for ii, word in int_to_vocab.items()}
      
    # return tuple
    return (vocab_to_int, int_to_vocab)

"""
DON'T MODIFY ANYTHING IN THIS CELL THAT IS BELOW THIS LINE
"""
tests.test_create_lookup_tables(create_lookup_tables)
````

### Tokenize Punctuation
We'll be splitting the script into a word array using spaces as delimiters.  However, punctuations like periods and exclamation marks can create multiple ids for the same word. For example, "bye" and "bye!" would generate two different word ids.

Implement the function `token_lookup` to return a dict that will be used to tokenize symbols like "!" into "||Exclamation_Mark||".  Create a dictionary for the following symbols where the symbol is the key and value is the token:
- Period ( **.** )
- Comma ( **,** )
- Quotation Mark ( **"** )
- Semicolon ( **;** )
- Exclamation mark ( **!** )
- Question mark ( **?** )
- Left Parentheses ( **(** )
- Right Parentheses ( **)** )
- Dash ( **-** )
- Return ( **\n** )

This dictionary will be used to tokenize the symbols and add the delimiter (space) around it.  This separates each symbols as its own word, making it easier for the neural network to predict the next word. Make sure you don't use a value that could be confused as a word; for example, instead of using the value "dash", try using something like "||dash||".

````python
def token_lookup():
    """
    Generate a dict to turn punctuation into a token.
    :return: Tokenized dictionary where the key is the punctuation and the value is the token
    """
    # TODO: Implement Function
    
    punc_token_dict = {
                        '.': '||period||',
                        ',': '||comma||',
                        '"': '||quotation_mark||',
                        ';': '||semicolon',
                        #':': '||colon||',
                        '!': '||exclamation_mark||',
                        '?': '||question_mark||',
                        '(': '||left_paren||',
                        ')': '||right_paren||',
                        '-': '||dash||',
                        '\n': '||new_line||'
                      }
         
    return punc_token_dict

"""
DON'T MODIFY ANYTHING IN THIS CELL THAT IS BELOW THIS LINE
"""
tests.test_tokenize(token_lookup)
````

````python
# Here we turn the words in the scripts into tokenized integers and save the results in a file called 'script_in_integers'
# By saving it to a file, we should save a lot of time while debugging the next step on code because we won't have to 
#  re-generate this data over and over again.

from collections import Counter
import pickle

punc_token_dict = token_lookup()
#print(len(punc_token_dict))
updated_text = []
#print(updated_text)


def encode_words(words):
    
    #print(words[0:10])
    
    # NOTE TO SELF: We may need to strip off words that are rarely used to reduce noise.
    
    # Create a dictionary that maps words to integers
    counts = Counter(words)                                         # the "key" = word, 
                                                                    # the "value" = number of times the word happened
    #print(counts['the'])
    #print(counts['||period||'])
    # Sort the counts dictionary from most used word to least used word
    vocab = sorted(counts, key=counts.get, reverse=True)
    #print(vocab[0:10])
    vocab_to_int = {word: ii for ii, word in enumerate(vocab, 1)}   # start enumerating with the value of "one"
                                                                    # in case we need to pad with "zero"
    #print(vocab_to_int)
    
    #print(vocab_to_int['jerry:'])
    #print(vocab_to_int['do'])
    #print(vocab_to_int['||question_mark||'])
    
    # Now, we use "vocab_to_int" dictionary to tokenize each word in the scripts (called words) 
    script_ints = []                                                # This is our tokenized script
    for word in words:
        script_ints.append(vocab_to_int[word])
        
    #print(script_ints[:10])
    #print(type(script_ints))
        
    return script_ints


def insert_punc_token(word_num_location, punc_location, word_lst):
    
    word = word_lst[word_num_location]
    #print(word)
    char_lst = list(word)
    #print(char_lst)
    word_length = (len(word) - 1)
    #print(word_length)
    
    # Walk through each character in the word and see if it is a puncuation
    for x in range(len(char_lst)):
        #print(x)
        #print(punc_location)
        
        # If it is a punctuation, replace the puncuation with the proper punctuation token
        for p in punc_location:
            if x == p:
                #print("a char is the punc")
                #print(char_lst[x])
                char_lst[x] = " " + punc_token_dict[char_lst[x]] + " "
                #print(char_lst)
    
    word = "".join(char_lst)
    #print(word)
    
    return word
   

def find_punc(text, punc_token_dict):
    
    punc_location = []
    count = 0
    word_lst = text.split()
    word_lst_length = len(word_lst)
    #print(word_lst_length)
    #print(word_lst[count: 25])
    
    for word_num in word_lst[0: word_lst_length]:
        char_lst = list(word_num)
        num_chars = len(char_lst)
        word_num_location = count
        #print(word_num)
        #print(char_lst)
        #print(num_chars)
        #print(count)
        
        # Now, we walk through each letter in a word and see if it is a puncuation or not
        # If we do, we at the puncuation location to our list called "punc_location"
        for char_num in range(num_chars):
            #print(char_lst[char_num])
            #print(char_num)
            for i in punc_token_dict:
                #print(i)  # This is the actual 'key' into the punc_token_dict
                #print(punc_token_dict[i]) # This is the puncuation token we want to stuff into the text.
                if (i == char_lst[char_num]):
                    #print('This is the offending puncuation mark \"', char_lst[char_num], '\"')
                    # We've identified a puncuation mark in the word at location "char_num"
                    punc_location.append(char_num)
                    #print(punc_location)
                    
        # Now we have the locations of puncuations in the word.  punc_location
        # Pass punch_location and the word itself to a routine that will stuff the puncuation token in
        # The word is located in word_lst at (count + word_num)       
        word = insert_punc_token(word_num_location, punc_location, word_lst)
        #print(word)
        word_lst[word_num_location] = word
        
        #print(count)
        count += 1
        punc_location = []
        
                    
    #print(word_lst[0: 25])

    return word_lst

def fix_punc_in_data(text, punc_token_dict):
    
    word_lst = find_punc(text, punc_token_dict)
    # Now pull it all back together like the original text, but with the puncuation tokens instead
    updated_text = " ".join(word_lst)
    # Now we can split it all back apart into words, which includes aour puncuation tokens
    #updated_text = updated_text.split()
    #print(updated_text[0: 25])
    #print(updated_text[88: 103])
    #print(updated_text[:500])
    words = updated_text.split()
    
    #encode_words(words)
    
    return words

words = fix_punc_in_data(text, punc_token_dict)
script_ints = encode_words(words)

script_ints_file = open('script_in_integers', 'ab')
pickle.dump(script_ints, script_ints_file)
script_ints_file.close()


#print(type(script_ints))
#print(script_ints[:100])
````

## Neural Network Training

With the structure of the network complete and data ready to be fed in the neural network, it's time to train it.

### Train Loop

The training loop is implemented for you in the `train_decoder` function. This function will train the network over all the batches for the number of epochs given. The model progress will be shown every number of batches. This number is set with the `show_every_n_batches` parameter. You'll set this parameter along with other parameters in the next section.

````python
"""
DON'T MODIFY ANYTHING IN THIS CELL
"""

def train_rnn(rnn, batch_size, optimizer, criterion, n_epochs, show_every_n_batches=100):
    batch_losses = []
    
    rnn.train()

    print("Training for %d epoch(s)..." % n_epochs)
    for epoch_i in range(1, n_epochs + 1):
        
        # initialize hidden state
        hidden = rnn.init_hidden(batch_size)
        
        for batch_i, (inputs, labels) in enumerate(train_loader, 1):
            
            # make sure you iterate over completely full batches, only
            n_batches = len(train_loader.dataset)//batch_size
            if(batch_i > n_batches):
                break
            
            # forward, back prop
            loss, hidden = forward_back_prop(rnn, optimizer, criterion, inputs, labels, hidden)          
            # record loss
            batch_losses.append(loss)

            # printing loss stats
            if batch_i % show_every_n_batches == 0:
                print('Epoch: {:>4}/{:<4}  Loss: {}\n'.format(
                    epoch_i, n_epochs, np.average(batch_losses)))
                batch_losses = []

    # returns a trained rnn
    return rnn
````

## Generate TV Script
With the network trained and saved, you'll use it to generate a new, "fake" Seinfeld TV script in this section.

### Generate Text
To generate the text, the network needs to start with a single word and repeat its predictions until it reaches a set length. You'll be using the `generate` function to do this. It takes a word id to start with, `prime_id`, and generates a set length of text, `predict_len`. Also note that it uses topk sampling to introduce some randomness in choosing the most likely next word, given an output set of word scores!

````python
"""
DON'T MODIFY ANYTHING IN THIS CELL THAT IS BELOW THIS LINE
"""
import torch.nn.functional as F

def generate(rnn, prime_id, int_to_vocab, token_dict, pad_value, predict_len=100):
    """
    Generate text using the neural network
    :param decoder: The PyTorch Module that holds the trained neural network
    :param prime_id: The word id to start the first prediction
    :param int_to_vocab: Dict of word id keys to word values
    :param token_dict: Dict of puncuation tokens keys to puncuation values
    :param pad_value: The value used to pad a sequence
    :param predict_len: The length of text to generate
    :return: The generated text
    """
    rnn.eval()
    
    # create a sequence (batch_size=1) with the prime_id
    current_seq = np.full((1, sequence_length), pad_value)
    current_seq[-1][-1] = prime_id
    predicted = [int_to_vocab[prime_id]]
    
    for _ in range(predict_len):
        if train_on_gpu:
            current_seq = torch.LongTensor(current_seq).cuda()
        else:
            current_seq = torch.LongTensor(current_seq)
        
        # initialize the hidden state
        hidden = rnn.init_hidden(current_seq.size(0))
        
        # get the output of the rnn
        output, _ = rnn(current_seq, hidden)
        
        # get the next word probabilities
        p = F.softmax(output, dim=1).data
        if(train_on_gpu):
            p = p.cpu() # move to cpu
         
        # use top_k sampling to get the index of the next word
        top_k = 5
        p, top_i = p.topk(top_k)
        top_i = top_i.numpy().squeeze()
        
        # select the likely next word index with some element of randomness
        p = p.numpy().squeeze()
        word_i = np.random.choice(top_i, p=p/p.sum())
        
        # retrieve that word from the dictionary
        word = int_to_vocab[word_i]
        predicted.append(word)     
        
        # the generated word becomes the next "current sequence" and the cycle can continue
        current_seq = np.roll(current_seq, -1, 1)
        current_seq[-1][-1] = word_i
    
    gen_sentences = ' '.join(predicted)
    
    # Replace punctuation tokens
    for key, token in token_dict.items():
        ending = ' ' if key in ['\n', '(', '"'] else ''
        gen_sentences = gen_sentences.replace(' ' + token.lower(), key)
    gen_sentences = gen_sentences.replace('\n ', '\n')
    gen_sentences = gen_sentences.replace('( ', '(')
    
    # return all the sentences
    return gen_sentences
````
